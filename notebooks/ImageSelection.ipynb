```python
import pandas as pd
import os
import shutil
import random
from collections import Counter
```

```python
image_dir = "../dataset/bird_images/nabirds/images"
final_dir = "../dataset/bird_images/nabirds/final_images"
mapping_file = "../dataset/bird_images/nabirds/classes.txt"
```

```python
mapping_df = pd.read_csv(mapping_file, delimiter = '\t', names = ['birds'], skiprows=1)
```

```python
mapping_df['folder'] = mapping_df.index + 1
```

```python
mapping_df['birds'] = mapping_df['birds'].str.split(n=1).str[1]
```

```python
mapping_df['birds'] = mapping_df['birds'].apply(lambda x: x.split('(')[0].strip())
```

```python
mapping_df
```

```python
print('Converting DataFrame to Dictionary')
bird_mapping = mapping_df.set_index('folder')['birds'].to_dict()
```

```python
#Iterate over each folder in the image directory
count = 0
for folder in os.listdir(image_dir):
    count+=1
    print(str(count) + ' - ' +folder)
    folder_path = os.path.join(image_dir, folder)
    
    images = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
    #select a random image
    random_image = random.choice(images)
    #Get the bird name from mapping
    target_folder = folder.lstrip('0')
    bird_name = bird_mapping.get(int(target_folder))
    #Rename the image and move to destination
    new_name = bird_name + os.path.splitext(random_image)[1]
    final_path = os.path.join(final_dir, new_name)
    shutil.copyfile(os.path.join(folder_path, random_image), final_path)

    print(f"Folder '{folder}' Renamed '{random_image}' to '{new_name}' and moved to '{final_dir}'")

print("Task completed!")
```
