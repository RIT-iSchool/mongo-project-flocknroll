```python
import requests
from bs4 import BeautifulSoup
import os

import pandas as pd
```

```python
birds_file = "/home/student/610_production_code/mongo-project-flocknroll/dataset/birds/birds.csv"
folder_path = "/home/student/610_production_code/mongo-project-flocknroll/dataset/bird_images/nabirds/new_images"
```

```python

def download_bird_image(bird_name, american_name, folder_path):

    try:
        url = f"https://ebird.org/species/{bird_name}"
        response = requests.get(url)
        response.raise_for_status()  

        soup = BeautifulSoup(response.content, "html.parser")

        image_elements = soup.find_all("img")

        if image_elements:
            image_element = image_elements[0]
            image_url = image_element["src"]
            image_response = requests.get(image_url, stream=True)
            image_response.raise_for_status() 

            image_filename = f"{american_name}.jpg"
            image_path = os.path.join(folder_path, image_filename)

            try:
                with open(image_path, "wb") as f:
                    for chunk in image_response.iter_content(1024):
                        f.write(chunk)
            except:
                print(f"skipped due to bad american name for {bird_name} to {image_path}")

            print(f"Downloaded image for {bird_name} to {image_path}")
        else:
            print(f"Image not found for {bird_name}")

    except:
        print(f"Error downloading image for {bird_name}")


```

```python
df = pd.read_csv(birds_file)
```

```python
species_codes = df["species_code"].to_list()[82:]
american_names = df["american_english_name"].to_list()[82:]
```

```python
# Replace with your bird names list and desired folder path
# bird_names = ["robin", "blue jay", "cardinal"]
# folder_path = "bird_images"

# Create the folder if it doesn't exist
os.makedirs(folder_path, exist_ok=True)

for bird_name, american_name in zip(species_codes, american_names):
    download_bird_image(bird_name, american_name, folder_path)
```
